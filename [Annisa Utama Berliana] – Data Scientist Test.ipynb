# import library

```python
import requests
import pandas as pd
import numpy as np
pd.set_option('display.max_rows',None,'display.max_columns',None,'display.max_colwidth',800)
```

# load data

```python
url = 'https://raw.githubusercontent.com/indrasetiadhip/data-task-sample/main/data-sample.json'
f = requests.get(url)
data = f.json()
```

```python
dataFrame = pd.DataFrame(columns=data[0].keys())
for i in range(len(data)):
    dataFrame.loc[i] = data[i].values()
    print(dataFrame)
```

```python
dataFrame.info()
```

```python
dataFrame['taskLocationDone'] = dataFrame['taskLocationDone'].astype(str).str.replace('{','').str.replace('}','').str.replace("'","")
dataFrame['cod'] = dataFrame['cod'].astype(str).str.replace('{','').str.replace('}','').str.replace("'","")
dataFrame['UserVar'] = dataFrame['UserVar'].astype(str).str.replace('{','').str.replace('}','').str.replace("'","")
dataFrame[['taskLocationDone_lon','taskLocationDone_lat']]=dataFrame['taskLocationDone'].str.split(pat=',',
                                                                                                   n=1, expand = True)
dataFrame[['cod_amount','cod_received']]=dataFrame['cod'].str.split(pat=',',
                                                            n=1, expand = True)
dataFrame[['UserVar_branch_dest','UserVar_taskStatusLabel']]=dataFrame['UserVar'].str.split(pat=',',
                                                            n=1, expand = True)
dataFrame[['UserVar_taskStatusLabel','UserVar_receiver_city']]=dataFrame['UserVar_taskStatusLabel'].str.split(pat=',',
                                                            n=1, expand = True)
dataFrame[['UserVar_receiver_city','UserVar_taskStatus']]=dataFrame['UserVar_receiver_city'].str.split(pat=', taskStatus:',
                                                            n=2, expand = True)
dataFrame[['UserVar_receiver_city','UserVar_branch_origin']]=dataFrame['UserVar_receiver_city'].str.split(pat=', branch_origin:',
                                                            n=2, expand = True)
dataFrame[['UserVar_receiver_city','UserVar_weight']]=dataFrame['UserVar_receiver_city'].str.split(pat=', weight:',
                                                            n=2, expand = True)
dataFrame[['UserVar_receiver_city','UserVar_taskDetailStatus']]=dataFrame['UserVar_receiver_city'].str.split(pat=', taskDetailStatus:',
                                                            n=2, expand = True)
dataFrame[['UserVar_receiver_city','UserVar_taskDetailStatusLabel']]=dataFrame['UserVar_receiver_city'].str.split(pat=', taskDetailStatusLabel:',
                                                            n=2, expand = True)
dataFrame['taskLocationDone_lon'] = dataFrame['taskLocationDone_lon'].str.replace('lon:','')
dataFrame['taskLocationDone_lat'] = dataFrame['taskLocationDone_lat'].str.replace('lat:','')
dataFrame['cod_amount'] = dataFrame['cod_amount'].str.replace('amount:','')
dataFrame['cod_received'] = dataFrame['cod_received'].str.replace('received:','')
dataFrame['UserVar_branch_dest'] = dataFrame['UserVar_branch_dest'].str.replace('branch_dest:','')
dataFrame['UserVar_taskStatusLabel'] = dataFrame['UserVar_taskStatusLabel'].str.replace('taskStatusLabel:','')
dataFrame['UserVar_receiver_city'] = dataFrame['UserVar_receiver_city'].str.replace('receiver_city:','')
dataFrame_ = dataFrame.drop(columns=['taskLocationDone','cod','UserVar'],axis=1)
```

```python
dataFrame_['UserVar_taskDetailStatusLabel'] = np.where(dataFrame_['UserVar_taskDetailStatus'].str.contains('YANG BERSANGKUTAN'),
                                                      'YANG BERSANGKUTAN',
                                                       np.where(dataFrame_['UserVar_taskDetailStatus'].str.contains('KELUARGA/SAUDARA'),
                                                               'KELUARGA/SAUDARA',dataFrame_['UserVar_taskDetailStatusLabel']))
dataFrame_['UserVar_taskDetailStatus'] = dataFrame_['UserVar_taskDetailStatus'].str.replace(', taskDetailStatusLabel: YANG BERSANGKUTAN','')
dataFrame_['UserVar_taskDetailStatus'] = dataFrame_['UserVar_taskDetailStatus'].str.replace(', taskDetailStatusLabel: KELUARGA/SAUDARA','')
print(dataFrame_.shape)
dataFrame_.head(2)
```

# insight

```python
# banyak paket yang diterima langsung oleh penerima atau selain penerima
taskDetailStatusLabel_map = {' YANG BERSANGKUTAN':'YANG BERSANGKUTAN',
                            'YANG BERSANGKUTAN':'YANG BERSANGKUTAN',
                            ' KELUARGA/SAUDARA':'KELUARGA/SAUDARA',
                            'KELUARGA/SAUDARA':'KELUARGA/SAUDARA'}
insight_1 = dataFrame_.copy()
insight_1['UserVar_taskDetailStatusLabel'] = insight_1['UserVar_taskDetailStatusLabel'].map(taskDetailStatusLabel_map )
insight_1['UserVar_taskDetailStatusLabel'] = insight_1['UserVar_taskDetailStatusLabel'].str.capitalize()
print(f'Jumlah paket yang diterima oleh penerima vs bukan penerima : \n',insight_1['UserVar_taskDetailStatusLabel'].value_counts())
insight_1['UserVar_taskDetailStatusLabel'].value_counts().plot(kind='pie',legend = True)
```

```python
# durasi lama pengiriman setiap agent jasa pengiriman barang
insight_2 = dataFrame_.loc[dataFrame_['flow']=='done']
insight_2['taskCreatedTime'] = pd.to_datetime(insight_2['taskCreatedTime'])
insight_2['taskCompletedTime'] = pd.to_datetime(insight_2['taskCompletedTime'])
insight_2['durasi_pengiriman'] = insight_2['taskCompletedTime']-insight_2['taskCreatedTime']
insight_2['durasi_in_hour'] = round(insight_2['durasi_pengiriman']/pd.Timedelta(hours=1))
insight_2['durasi_in_minutes'] = round(insight_2['durasi_pengiriman']/pd.Timedelta(minutes=1))
insight_2['durasi_in_seconds'] = round(insight_2['durasi_pengiriman']/pd.Timedelta(seconds=1))
insight_2['bin_duration_in_hour'] = pd.qcut(insight_2['durasi_in_hour'],4)
insight_2.groupby(['bin_duration_in_hour','taskAssignedTo']).agg({'taskId':['count']})
```

```python
# berdasarkan tabel diatas dapat kita lihat :
# 1. jasa pengiriman barang 'cautiousCoconut1' dapat disarankan sebagai jasa pengiriman barang dengan durasi pengiriman
# yang tercepat dibandingkan jasa pengiriman lainnya. terlihat dari banyaknya orderan yang sampai dalam waktu kurang dari 3 jam
# 2. jasa pengiriman barang 'crummyCrane8' adalah jasa pengiriman barang yang paling lama yaitu antara 7 sampai 11 jam.
```

```python
# biaya pengiriman barang setiap agen jasa pengirimn per kg
insight_3 = dataFrame_.copy()
insight_3['cod_amount'] = insight_3['cod_amount'].astype(int)
insight_3['UserVar_weight'] = insight_3['UserVar_weight'].astype(float)
insight_3 = insight_3.groupby(['taskAssignedTo']).agg({'cod_amount':'mean',
                                          'UserVar_weight':'mean'}).reset_index()
insight_3['cod_amount_perkg']=round(insight_3['cod_amount'] /insight_3['UserVar_weight'])
insight_3.sort_values(by=['cod_amount_perkg'],ascending=True)
```

```python
# dari tabel diatas dapat kita lihat:
# humorousSausage2 merupakan jasa pengiriman dengan biaya pengiriman per kg termahal
# crummyCrane8 merupakan jasa pengiriman dengan biaya pengiriman per kg termurah
```

```python
# lokasi pengiriman barang tersering dan terjarang
insight_4 = dataFrame_.copy()
print(f'jumlah pengiriman barang dari setiap branch origin : \n',insight_4['UserVar_branch_origin'].value_counts())
insight_4['UserVar_branch_origin'].value_counts().plot(kind='pie',legend = True)
```

```python
print(f'jumlah penerimaan barang dari setiap branch dest : \n',insight_4['UserVar_branch_dest'].value_counts())
insight_4['UserVar_branch_dest'].value_counts().plot(kind='pie',legend = True)
```

```python
insight_4 = insight_4.groupby(['UserVar_branch_origin','UserVar_branch_dest']).agg({'taskId':'count'}).reset_index()
insight_4.sort_values(by=['UserVar_branch_origin','taskId'],ascending=[False,False])
```

```python
# dari situ dapat kita lihat:
# branch origin terbanyak adalah CGK
# brach destination terbanyak adalah PDG
# namun secara keseluruhan rute pengiriman yang paling sering terjadi adalah pada CGK ke CLG
```
